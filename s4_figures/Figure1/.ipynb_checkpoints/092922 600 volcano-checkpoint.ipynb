```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pipe_util2
import os
```

# master sheet processing

```python
# master sheet
master_file = "Master_spreadsheet_2.csv"
master = pd.read_csv(master_file, header=0)
master.head(3)
```

```python
master.columns
```

```python
master_600_150d6 = master[['UniqueID', 'Symbol',
                           'P-value (150 vs 600D6)', 'FDR.2', '150 Mean',
                           '600D6 Mean', 'FC.2']].copy()
master_600_150d6.head(3)
```

```python
master_600_150d6["FC2"] = master_600_150d6["600D6 Mean"]/master_600_150d6["150 Mean"]
master_600_150d6["FC2log"] = np.log2(master_600_150d6["FC2"])
master_600_150d6["logp"] = -np.log10(master_600_150d6["P-value (150 vs 600D6)"])
master_600_150d6.head(3)
```

```python
master_600_150d6.columns
```

```python
master_600_150d6_simple = master_600_150d6[['UniqueID', 'Symbol', 'P-value (150 vs 600D6)', 
                                                'FDR.2', 'FC2', 'FC2log', "logp"]].copy()
master_600_150d6_simple.head(3)
```

```python
master_600_150d6_simple.shape
```

```python
master_600_150d6_simple.dropna(axis = 0, how = "any", 
                               subset=["Symbol"], 
                               inplace = True)
master_600_150d6_simple.drop_duplicates(inplace=True, 
                                        ignore_index=True)
master_600_150d6_simple.shape
```

```python
def df_merge(master, file):
    
    # Go process name
    name = os.path.basename(file)[3:-4]
    name = name.replace("_", " ")
    name = name.capitalize()
    
    df = pd.read_csv(file, sep = "\t", header = 0, index_col=0)
    # merge dataframe
    df_merge = pd.merge(df, master, 
                        how = "left", 
                        left_on = "GENE SYMBOL", right_on = "Symbol")
    # remove duplicated genes
    df_merge.sort_values(["Symbol", "FC2log"], ascending=False, 
                         inplace=True)
    df_merge_simple = df_merge.drop_duplicates(subset="Symbol").copy()
    # selecting few columns
    df_merge_simple = df_merge_simple[['UniqueID', 'Symbol', 
                                       'GENE_TITLE', 
                                       'P-value (150 vs 600D6)', 
                                       'FDR.2', 'FC2','FC2log', 
                                       'logp']].copy()
    # go term name
#     df_merge_simple["go"] = name
    # reset index
    df_merge_simple.reset_index(drop=True, inplace=True)
    return df_merge_simple
```

```python
df_list = []
for f in files:
    df = df_merge(master_600_150d6_simple, f)
    df_list.append(df)
df_all = pd.concat(df_list)
df_all.drop_duplicates(inplace=True)
df_all.sort_values(["FC2"], ascending=False, inplace=True)
df_all.reset_index(drop=True, inplace=True)
df_all.shape
```

```python
plt.figure(figsize=(4,6))
# plt.scatter(df_all.FC2log, df_all.logp, s=12)

p_thre = -np.log10(0.05)

df_all_up = df_all[(df_all.FC2log>1) & (df_all.logp>p_thre)].copy()
up_num = df_all_up.shape[0]
plt.scatter(df_all_up.FC2log, df_all_up.logp, 
            s=12, c="red", label = "Up regulated: %s"%(up_num))
df_all_down = df_all[(df_all.FC2log<-1) & (df_all.logp>p_thre)].copy()
down_num = df_all_down.shape[0]
plt.scatter(df_all_down.FC2log, df_all_down.logp, 
            s=12, c="blue", label = "Down regulated: %s"%(down_num))
df_all_non = df_all[(df_all.FC2log.abs()<=1) | (df_all.logp<=p_thre)].copy()
non_num = df_all_non.shape[0]
plt.scatter(df_all_non.FC2log, df_all_non.logp, s=12, c="gray")


plt.axvline(x = -1,c="k", linestyle="-.")
plt.axvline(x = 1, c="k", linestyle="-.")
plt.axhline(y = p_thre, c="k", linestyle="-.")

# label top genes
df_all_top = pd.concat([df_all_up, df_all_down])
df_all_top["FC2logabs"] = df_all_top["FC2log"].abs()
df_all_top.sort_values(["FC2log"], ascending=False, inplace = True)
df_all_top_save = df_all_top.copy()
df_all_top_save.to_csv("Volcano_data_600.csv") # save to file
df_all_top.sort_values(["FC2logabs"], ascending=False, inplace = True)
print(df_all_top.shape[0])
df_all_top = df_all_top.loc[:15, :]
for i in range(df_all_top.shape[0]):
    plt.text(df_all_top.FC2log.values[i]+0.05, df_all_top.logp.values[i]+0.05, 
            str(df_all_top.Symbol.values[i]), 
            color='k', fontsize=7)
    
df_all_non.sort_values(["FC2log"], ascending=False, inplace = True)
df_all_non.to_csv("Volcano_non_data.csv")

plt.legend(title="Differential expressed genes", 
           bbox_to_anchor=(1, 1.18),
           frameon=True)
    
# plt.xlim((-3.5, 3.5))
plt.xlabel(r"$log_2(fold\ change)$", fontsize=12)
plt.ylabel(r"$-log_{10}(pval)$", fontsize=12)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(True)
plt.gca().spines['left'].set_visible(True)

plt.tight_layout()
plt.savefig("Volcano_600.pdf")
```
